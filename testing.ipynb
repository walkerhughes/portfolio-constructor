```python
import yfinance as yf
import pandas as pd 
```

```python
def get_sector_etf_tickers(): 
    return ["XLC", "XLF", "XLRE", "XLY", "XLV", "XLK", "XLE", "XLI", "XLU", "XLP", "XLB"]
```

```python
df = yf.download(tickers=get_sector_etf_tickers(), start="2022-12-31", end="2023-12-31").reset_index()
```
